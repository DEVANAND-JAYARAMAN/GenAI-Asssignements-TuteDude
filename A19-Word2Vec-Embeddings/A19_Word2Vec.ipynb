# Task 1: Understanding Word Embeddings

### 1. What are word embeddings?

Word embeddings are dense numerical vector representations of words where words with similar meanings are placed closer together in a multidimensional vector space.

---

### 2. Why do One-Hot Encoding and Bag of Words fail?

One-Hot Encoding and Bag of Words represent words independently without understanding their meanings or relationships. They create sparse vectors and cannot capture semantic similarity or context.

---

### 3. How do word embeddings solve these problems?

Word embeddings learn meaningful numerical representations from the context in which words appear. Similar words receive similar vectors, allowing NLP models to understand semantic relationships between words.

# Task 2: Word2Vec Overview

### What is Word2Vec?

Word2Vec is a shallow neural network model that learns vector representations of words by predicting words from their surrounding context. It captures semantic and syntactic relationships between words.

---

### Predicting Words from Context

Instead of counting words, Word2Vec learns by observing neighboring words. During training, it predicts either the target word from surrounding words (CBOW) or surrounding words from the target word (Skip-Gram).

---

### Vocabulary

The vocabulary is the collection of all unique words present in the training corpus.

---

### Context Window

The context window defines how many neighboring words on both sides are considered while learning word relationships.

Example:

I love machine learning

Window = 2

Context for "machine":

love, learning

---

### Embedding Dimension

Embedding dimension is the length of the numerical vector representing each word.

Example:

Word → Vector

movie → [0.14, -0.81, 0.32, ...]

In this assignment we use:

Embedding Size = 100

# Task 3: CBOW vs Skip-Gram

## CBOW (Continuous Bag of Words)

CBOW predicts the target word using the surrounding context words. It trains faster and performs well on larger datasets where common words appear frequently.

Example:

I love ____ learning

Prediction: machine

CBOW is generally preferred when the dataset is large and training speed is important.

---

## Skip-Gram

Skip-Gram predicts the surrounding context words using the target word. It is slower than CBOW but produces better representations for rare words.

Example:

Target Word:

machine

Predicted Words:

love, learning

Skip-Gram is useful when the dataset contains many uncommon or infrequent words.


# Task 4: Neural Network Intuition Behind Word2Vec

## Input Layer

Each input word is represented as a one-hot encoded vector.

---

## Hidden Layer (Embedding Layer)

The hidden layer converts the sparse one-hot vector into a dense embedding vector. These learned weights become the word embeddings.

---

## Output Layer

The output layer predicts either the target word (CBOW) or the surrounding context words (Skip-Gram).

---

## How Weights Become Word Embeddings

During training, the neural network updates its weights to improve prediction accuracy. After training, these learned weights form meaningful vector representations of words.

```python
#Task 5: Prepare Text for Word2Vec

!pip install gensim

#error came and checked that I actually missed to install the gensim
import pandas as pd
import numpy as np

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

df = pd.read_csv("/content/processed_sms_dataset.csv")
df.head()
```

```python
# Convert every sentence into a list of words
sentences = []
for text in df["final_clean_text"]:
    tokens = word_tokenize(str(text))
    sentences.append(tokens)

print("First 5 Tokenized Sentences:\n")
for sentence in sentences[:5]:
    print(sentence)

print(type(sentences))
print(type(sentences[0]))
print(type(sentences[0][0]))
```

```python
#Task 6 – Train Word2Vec (CBOW)

cbow_model = Word2Vec(sentences=sentences,vector_size=100,window=5,min_count=1,sg=0)
print("Vocabulary Size:")
print(len(cbow_model.wv.index_to_key))

print(cbow_model.wv.index_to_key[:20])
sample_word = cbow_model.wv.index_to_key[0]
print("Sample Word:", sample_word)
print(cbow_model.wv[sample_word])

print(cbow_model.wv[sample_word].shape)
cbow_model.save("cbow_word2vec.model")
```

```python
#Task 7 – Train Skip-Gram Word2Vec Model
import time

start_time = time.time()
skipgram_model = Word2Vec(sentences=sentences,vector_size=100,window=5,min_count=1,sg=1)
end_time = time.time()
skipgram_time = end_time - start_time

print(f"Skip-Gram Training Time: {skipgram_time:.4f} seconds")

#comparing with the CBOW

start_time = time.time()
cbow_model = Word2Vec(sentences=sentences,vector_size=100,window=5,min_count=1,sg=0)
end_time = time.time()
cbow_time = end_time - start_time
print(f"CBOW Training Time: {cbow_time:.4f} seconds")

#comparing the training time using this code:
comparison_df = pd.DataFrame({"Model": ["CBOW", "Skip-Gram"],"Training Time (seconds)": [cbow_time, skipgram_time]})
comparison_df

print(cbow_model.wv.index_to_key[:30])
word = "free"

print("CBOW Similar Words:\n")
print(cbow_model.wv.most_similar(word, topn=5))

print("Skip-Gram Similar Words:\n")
print(skipgram_model.wv.most_similar(word, topn=5))
```

```python
#Task 8 – Word Similarity & Vector Operations
sample_word = "free"
similar_words = cbow_model.wv.most_similar(sample_word, topn=10)
print(f"Most Similar Words to '{sample_word}':\n")
for word, score in similar_words:
    print(f"{word:<15} {score:.4f}")

similarity = cbow_model.wv.similarity("free", "win")
print("Similarity between 'free' and 'win':")
print(similarity)

```

# Task 10: Observations & Limitations

## 1. Difference between CBOW & Skip-Gram

CBOW predicts a target word using surrounding context words and trains faster. Skip-Gram predicts surrounding words from a target word and generally performs better for infrequent words, though it requires more training time.


## 2. Advantages of Word2Vec over TF-IDF

Word2Vec captures semantic relationships and contextual similarities between words using dense vector representations. In contrast, TF-IDF represents words based only on their frequency without understanding meaning.

## 3. Limitations of Word2Vec

- Produces one vector per word regardless of context.
- Cannot distinguish different meanings of the same word.
- Requires sufficient training data for high-quality embeddings.
- Does not capture long-range context.

## 4. Why Context Matters in Modern NLP

The meaning of a word depends on the sentence in which it appears. Modern transformer models generate contextual embeddings, allowing the same word to have different representations based on its usage. This leads to more accurate language understanding.
