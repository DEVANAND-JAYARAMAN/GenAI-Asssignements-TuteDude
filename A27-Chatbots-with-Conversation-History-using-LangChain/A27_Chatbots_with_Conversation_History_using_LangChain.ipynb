Task 1 - Understanding Chat Messages

1. What are SystemMessage, HumanMessage, and AIMessage?

SystemMessage: Gives instructions or defines the behavior of the AI.
HumanMessage: Represents the user's input or question.

2. Why is message-based prompting better than a single prompt?

Message-based prompting keeps conversations organized by separating system instructions, user messages, and AI responses. It also helps the chatbot understand previous context, making conversations more natural and consistent.

Task 2 - Using MessagesPlaceholder

In this task, I created a chat prompt using ChatPromptTemplate and MessagesPlaceholder to include previous conversation history dynamically.

```python
!pip install -q langchain langchain-groq
import os

from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

#loading the api key from the colab secrets
from getpass import getpass
os.environ["GROQ_API_KEY"] = getpass("Enter your Groq API Key: ")

#creating the llm
llm = ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
print("Groq model loaded successfully")
```

```python
prompt = ChatPromptTemplate.from_messages([("system", "You are a helpful AI assistant."),
        MessagesPlaceholder(variable_name="chat_history"),
        ("human", "{question}")])

print("Prompt template created")
chat_history = [HumanMessage(content="What is Python?"),AIMessage(content="Python is a programming language.")]

chain = prompt | llm
response = chain.invoke({"chat_history": chat_history,"question": "Can you give me a simple example?"})
print("\nAI Response:\n")
print(response.content)
```

Observation

The chatbot successfully used MessagesPlaceholder to include previous conversation history. This helps the model understand the context before answering the current question.

Task 3 - Basic Message History

In this task, I stored the conversation history in a list and passed the entire history to the LLM for every user query.

```python
from langchain_core.messages import HumanMessage, AIMessage

# Store all messages here
chat_history = []
print("Simple Q&A Chatbot (type 'exit' to stop)\n")

while True:
    question = input("You: ")
    if question.lower() == "exit":
        print("Chat ended.")
        break
    # Store user message
    chat_history.append(HumanMessage(content=question))

    # Generation of the response
    response = chain.invoke(
        {"chat_history": chat_history,"question": question})
    print("\nAI:", response.content)
    # Store AI response
    chat_history.append(AIMessage(content=response.content))
    print("\nTotal Messages:", len(chat_history))
    print("-" * 60)
print("\nObservation:")
print("- Chat history is stored after every interaction.")
print("- Previous messages are sent to the model for better context.")
```

```python
# Task 4 - Trimming Chat History

from langchain_core.messages import HumanMessage, AIMessage
chat_history = []
MAX_MESSAGES = 6
print("Chatbot with Trimmed History (type 'exit' to stop)\n")
while True:
    question = input("You: ")
    if question.lower() == "exit":
        print("Chat ended.")
        break
    chat_history.append(HumanMessage(content=question))

    # Keep only the latest messages
    if len(chat_history) > MAX_MESSAGES:
        chat_history = chat_history[-MAX_MESSAGES:]

    response = chain.invoke({"chat_history": chat_history,"question": question})
    print("\nAI:", response.content)
    chat_history.append(AIMessage(content=response.content))

    if len(chat_history) > MAX_MESSAGES:
        chat_history = chat_history[-MAX_MESSAGES:]
    print("\nCurrent History Size:", len(chat_history))
    print("-" * 60)

print("\nObservation:")
print("- Older messages are removed when the history becomes too large.")
print("- The chatbot still answers using the most recent conversation.")
```

```python
# Task 5 - Q&A Chatbot

from langchain_core.messages import HumanMessage, AIMessage

# Chat history
chat_history = []'wh'
MAX_MESSAGES = 6
print("========== Q&A Chatbot ==========")
print("Type 'exit' to stop.\n")

while True:
    question = input("You: ")
    if question.lower() == "exit":
        print("Goodbye!")
        break
    # Save user message
    chat_history.append(HumanMessage(content=question))

    # Trim old messages
    if len(chat_history) > MAX_MESSAGES:
        chat_history = chat_history[-MAX_MESSAGES:]

    # Get response
    response = chain.invoke({"chat_history": chat_history,"question": question})
    print("\nAI:", response.content)
    # Save AI response
    chat_history.append(AIMessage(content=response.content))
    # Trim again after AI reply
    if len(chat_history) > MAX_MESSAGES:
        chat_history = chat_history[-MAX_MESSAGES:]
    print("\nCurrent Messages:", len(chat_history))
    print("-" * 60)

print("\nObservation:")
print("- The chatbot remembers previous questions.")
print("- Follow-up questions use earlier conversation as context.")
print("- Old messages are removed to avoid very long history.")
```

Task 7 - Observations & Insights
1. Why is chat history important?

Chat history helps the chatbot remember previous conversations, making responses more relevant and natural

2. Trade-offs between long memory and performance

Long conversation history improves context but increases token usage, response time, and cost. Shorter history is faster but may lose important context

3. When should we summarize instead of trimming history?

Summarization is useful for very long conversations because it preserves important information while reducing the number of tokens. Trimming simply removes older messages and may lose useful context

 4. Difference between MessagesPlaceholder and memory

MessagesPlaceholder inserts the conversation history into the prompt. Memory is used to store and manage previous messages. Together, they help maintain context during a conversation
