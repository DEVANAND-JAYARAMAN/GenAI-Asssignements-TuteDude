# Cloud Deployment

The original application was developed locally using Ollama and CodeLlama.

For Streamlit deployment, the application was modified to use the Groq API with the Llama 3.3 70B Versatile model because Streamlit Cloud cannot run a local Ollama server.

# Comparison

### Streamlit Community Cloud

Pros

- Easy deployment
- GitHub integration
- Secret management

Cons

- Cannot run Ollama locally

### Hugging Face Spaces

Pros

- AI project hosting
- Easy sharing

Cons

- Limited free resources

# When to use each platform

Use Streamlit Community Cloud for dashboards and web applications.

Use Hugging Face Spaces for showcasing AI and Machine Learning projects.
