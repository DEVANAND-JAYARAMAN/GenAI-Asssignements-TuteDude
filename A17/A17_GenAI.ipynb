### Common Issues Found in Raw Text

- Presence of uppercase and lowercase letters.
- Punctuation marks such as commas, periods and exclamation marks.
- Numbers present in some messages.
- Multiple spaces in a few messages.
- URLs and email addresses may appear.
- Special characters and symbols.


```python
#Task 2: Basic Text Cleaning
import string
import re

def basic_clean(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
#applying

df['clean_text_basic'] = df['text'].apply(basic_clean)

#comparing original vs cleaned text:

df[['text', 'clean_text_basic']].head(10)
```

```python
#Task 3: Removing Noise

def advanced_clean(text):

    # Convert to lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Remove punctuation and special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    return text.strip()
```

```python
#Apply the function
df["clean_text_advanced"] = df["text"].apply(advanced_clean)

#Comparing the results

df[["text", "clean_text_basic", "clean_text_advanced"]].head(10)
```

```python
nltk.download('punkt_tab')

#Task 4: Remove Stopwords

stop_words = set(stopwords.words("english"))
def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = []
    for word in words:
        if word not in stop_words:
            filtered_words.append(word)

    return " ".join(filtered_words)
#applying and comparing:

df["text_no_stopwords"] = df["clean_text_advanced"].apply(remove_stopwords)
df[["clean_text_advanced", "text_no_stopwords"]].head(10)
```

```python
#Task 5: Repeated Characters & Slang

slang_dict = {
    "u": "you",
    "ur": "your",
    "gr8": "great",
    "luv": "love",
    "msg": "message",
    "pls": "please",
    "thx": "thanks"
}

def normalize_text(text):

    # Replace repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    words = []

    for word in text.split():
        words.append(slang_dict.get(word, word))

    return " ".join(words)

#applying and comparing

df["normalized_text"] = df["text_no_stopwords"].apply(normalize_text)
df[["text_no_stopwords", "normalized_text"]].head(10)

```

```python
#Task 6: Tokenization

#Word Tokenization
print("Word Tokenization (First 3 Samples)\n")

for i in range(3):
    print(f"Sample {i+1}")
    print(word_tokenize(df.loc[i, "normalized_text"]))
    print("-"*60)

#Sentence Tokenization
print("Sentence Tokenization (First 3 Samples)\n")

for i in range(3):
    print(f"Sample {i+1}")
    print(sent_tokenize(df.loc[i, "text"]))
    print("-"*60)

```

```python
#Task 7: Stemming
stemmer = PorterStemmer()
sample_text = df.loc[0, "normalized_text"]

words = word_tokenize(sample_text)

stemmed_words = [stemmer.stem(word) for word in words]

comparison = pd.DataFrame({
    "Original Word": words,
    "Stemmed Word": stemmed_words
})

comparison
```

```python
#Task 8: Lemmatization

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

comparison = pd.DataFrame({
    "Original Word": words,
    "Stemmed": stemmed_words,
    "Lemmatized": lemmatized_words
})

comparison
```

```python
#Task 9: Final NLP Pipeline

def nlp_preprocess(text):
    # Advanced Cleaning
    text = advanced_clean(text)
    # Remove Stopwords
    text = remove_stopwords(text)
    # Normalize Text
    text = normalize_text(text)
    # Tokenization
    tokens = word_tokenize(text)
    # Lemmatization
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return " ".join(tokens)

#applying and showing the output

df["final_clean_text"] = df["text"].apply(nlp_preprocess)
df[["text", "final_clean_text"]].head(10)

#saving the processed dataset

df.to_csv("processed_sms_dataset.csv", index=False)

print("Processed dataset saved successfully.")
```

Task 10: Observations
Observations & Insights

1. Difference between Basic and Advanced Cleaning

Basic text cleaning removes common formatting issues such as uppercase letters, punctuation, numbers, and extra spaces. Advanced cleaning goes a step further by removing URLs, email addresses, HTML tags, and special characters, producing cleaner and more consistent text for NLP tasks.

2. Why Lemmatization is Preferred Over Stemming

Stemming simply removes word endings and may generate incomplete words such as studies → studi. Lemmatization uses vocabulary and language rules to convert words into their proper dictionary form, such as studies → study. This preserves the meaning of the text and is generally preferred for NLP applications.

3. Importance of Preprocessing in NLP Models

Text preprocessing removes unwanted noise, standardizes text, and reduces unnecessary variations in words. This helps NLP models focus on meaningful information, improves model accuracy, and reduces computational complexity during training.
