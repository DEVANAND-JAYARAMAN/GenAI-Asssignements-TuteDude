Task 1 - Ollama Setup & Basic Chat

In this task, I loaded the Ollama model using LangChain and tested it with a simple prompt

```python
# Load the Ollama model

llm = ChatOllama(model="llama3.2",temperature=0.3)
print("Model loaded successfully.")

@traceable(name="Ollama Chat")
def ask_llm(question):
    response = llm.invoke(question)
    return response.content

answer = ask_llm("What is LangChain?")
print(answer)
```

```python
question = "What is LangChain?"
with tracing_context(enabled=True):
    response = llm.invoke(question)
print(response.content)
```

```python
#basic prompt
question = "What is Generative AI? Explain in simple words."
response = llm.invoke(question)

print("Question:")
print(question)

print("\nResponse:")
print(response.content)
```
