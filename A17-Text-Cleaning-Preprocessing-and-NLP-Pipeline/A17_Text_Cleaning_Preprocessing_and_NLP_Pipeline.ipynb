# Task 1 - Understanding Raw Text Data

In this task, I created a small custom text dataset and inspected the raw text before preprocessing.

```python
import pandas as pd

texts = [
    "This product is AMAZING!!!",
    "Visit https://example.com for more details.",
    "Contact us at support@gmail.com",
    "<p>This is HTML content</p>",
    "I loooove this phone sooo much!!!",
    "The price is 999 rupees.",
    "  Extra    spaces   should be removed. ",
    "gr8 service, u should try it!",
    "Worst experience ever!!!",
    "Delivery was FAST and GOOD.",
    "Call me at 9876543210",
    "Weather is soooo hot today.",
    "Python is AWESOME!!!",
    "I can't believe this happened!!",
    "Special offer: Buy 2 Get 1 FREE.",
    "Nice movie",
    "Email me tomorrow.",
    "This is 100% genuine.",
    "Happy Birthday!!!",
    "Machine Learning is interesting.",
    "OpenAI develops AI models.",
    "Thanks a lot!!!!!",
    "This laptop costs 55000.",
    "HTML <b>Bold</b> text",
    "See you soon :)"
]

df = pd.DataFrame({"text": texts})
print(df.head())
print("\nLength of each text:\n")
print(df["text"].str.len())
```

```python
#to print all the cases, im going with all possible ways to print
print("\nCommon issues found in the dataset:")
print("-Uppercase letters")
print("-Punctuation")
print("-Numbers")
print("-Extra spaces")
print("-URLs")
print("-Email addresses")
print("-HTML tags")
print("-Emojis")
print("-Slang words")
print("-Repeated characters")
```

The raw dataset contains different types of noise such as uppercase letters, punctuation, numbers, URLs, HTML tags, emojis, extra spaces, repeated characters, and slang words. These need to be cleaned before NLP preprocessing

```python
# Task 2 - Basic Text Cleaning

import re
import string

def basic_clean(text):
    # Convert to lowercase
    text = text.lower()
    # Remove the punctuation marksers
    text = text.translate(str.maketrans("", "", string.punctuation))
    # Removing the numbers
    text = re.sub(r"\d+", "", text)
    # Removeing extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_text_basic"] = df["text"].apply(basic_clean)
print("Basic cleaning completed.\n")
```

```python
#comparing the original text and cleaned test
comparison = df[["text", "clean_text_basic"]]
print(comparison.head(10))


#random checking to verify
print("\nSample cleaned texts:\n")
for i in [3, 7, 12]:
    print(f"Original : {df.loc[i, 'text']}")
    print(f"Cleaned  : {df.loc[i, 'clean_text_basic']}")
    print("-" * 50)
```

### Observation from this task is

After basic cleaning: All text is converted to lowercase, Punctuation has been removed, numbers have been removed,and extra spaces have been normalized.

```python
#Task 3 – Removing Noise
import re

def advanced_clean(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    # Remove email addresses
    text = re.sub(r"\S+@\S+", "", text)
    # Remove HTML tags
    text = re.sub(r"<.*?>", "", text)
    # Remove emojis and other non-ASCII characters
    text = text.encode("ascii", "ignore").decode()
    # Remove special characters
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["clean_text_advanced"] = df["text"].apply(advanced_clean)
print("Advanced cleaning completed.")
```

```python
#Comparing the Results
df[["text", "clean_text_basic", "clean_text_advanced"]].head(10)

for i in [1, 2, 3, 8]:
    print(f"\nOriginal : {df.loc[i, 'text']}")
    print(f"Advanced : {df.loc[i, 'clean_text_advanced']}")
    print("-" * 50)
```

```python
# Task 4 - Handling Stopwords

import nltk
nltk.download("stopwords")

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

df["text_no_stopwords"] = df["clean_text_advanced"].apply(remove_stopwords)
print("Stopwords removed successfully.")
```

```python
#comapring beofre and after
df[["clean_text_advanced", "text_no_stopwords"]].head(10)

for i in [0, 5, 10]:
    print(f"\nBefore : {df.loc[i, 'clean_text_advanced']}")
    print(f"After  : {df.loc[i, 'text_no_stopwords']}")
    print("-" * 50)
```

```python
#Stopwords such as "is", "the", "at", and "for" were removed from the text. This helps reduce unnecessary words while preserving the important information for NLP tasks.
```

```python
# Task 5 - Handling Repeated Characters & Slang

import re
# Small slang dictionary
slang_dict = {"u": "you","gr8": "great","ur": "your","thx": "thanks","pls": "please"}
def normalize_text(text):
    words = text.split()
    cleaned_words = []

    for word in words:
        # Replace slang
        if word in slang_dict:
            word = slang_dict[word]
        # Normalize repeated characters (soooo -> soo, loooove -> loove)
        word = re.sub(r"(.)\1{2,}", r"\1\1", word)
        cleaned_words.append(word)
    return " ".join(cleaned_words)

df["normalized_text"] = df["text_no_stopwords"].apply(normalize_text)
print("Repeated characters and slang normalized.\n")

# Comparing before and after
for i in [4, 7, 11]:
    print(f"Original   : {df.loc[i, 'text']}")
    print(f"Before     : {df.loc[i, 'text_no_stopwords']}")
    print(f"Normalized : {df.loc[i, 'normalized_text']}")
    print("-" * 60)

print("\nObservation:")
print("- Slang words like 'u' and 'gr8' were replaced.")
print("- Repeated characters were normalized.")
print("- Text is now more consistent for further NLP preprocessing.")
```

```python
# Task 6 - Tokenization

import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download("punkt")
nltk.download("punkt_tab")
print("Tokenization Results\n")

for i in [0, 5, 10]:
    text = df.loc[i, "normalized_text"]
    print(f"Sample {i+1}")
    print("Text:", text)
    print("\nWord Tokens:")
    print(word_tokenize(text))
    print("\nSentence Tokens:")
    print(sent_tokenize(text))
    print("-" * 60)

print("\nObservation:")
print("- Word tokenization splits text into individual words.")
print("- Sentence tokenization splits text into complete sentences.")
```

```python
# Task 7 - Stemming

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
sample_words = ["running","studies","playing","better","happiness","connected","cars","easily"]
print("Original Word --> Stemmed Word\n")

for word in sample_words:
    print(f"{word:12} --> {stemmer.stem(word)}")

print("\nObservation:")
print("- Stemming removes word endings to produce the root form.")
print("- Sometimes the stemmed word is not a valid English word.")
```

```python
# Task 8 - Lemmatization

from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
nltk.download("omw-1.4")
lemmatizer = WordNetLemmatizer()
print("Original | Stemmed | Lemmatized\n")
for word in sample_words:
    stem = stemmer.stem(word)
    lemma = lemmatizer.lemmatize(word)
    print(f"{word:12} | {stem:10} | {lemma}")

print("\nObservation:")
print("- Lemmatization returns meaningful dictionary words.")
print("- Stemming is faster but may produce incomplete words.")
print("- Lemmatization is generally preferred for NLP tasks.")
```

```python
# Task 9 - Final NLP Pipeline

import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

slang_dict = {"u": "you","gr8": "great","ur": "your","thx": "thanks","pls": "please"}
def nlp_preprocess(text):
    # Lowercase
    text = text.lower()
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    # Remove emails
    text = re.sub(r"\S+@\S+", "", text)
    # Remove HTML tags
    text = re.sub(r"<.*?>", "", text)
    # Remove emojis / non-ascii characters
    text = text.encode("ascii", "ignore").decode()
    # Remove the punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    # Removing the numbers
    text = re.sub(r"\d+", "", text)
    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()
    # Tokenization
    words = word_tokenize(text)
    final_words = []
    for word in words:
        # Replace slang
        if word in slang_dict:
            word = slang_dict[word]
        # Normalize repeated characters
        word = re.sub(r"(.)\1{2,}", r"\1\1", word)
        # Remove the stopwords
        if word not in stop_words:
            word = lemmatizer.lemmatize(word)
            final_words.append(word)
    return " ".join(final_words)

# Apply pipeline
df["final_clean_text"] = df["text"].apply(nlp_preprocess)
print("Final NLP preprocessing completed.\n")

# Displaying is results
print(df[["text", "final_clean_text"]].head(10))

print("\nObservation:")
print("- The pipeline combines all preprocessing steps into one function.")
print("- The final text is clean and ready for NLP applications.")
```

Task 10 – Observations & Insights

1.Basic cleaning remove lowercase, punctuation, numbers and extra spaces. Advanced cleaning also remove URLs, emails, HTML tags, emojis and other unwanted things.

2. Lemmatization gives proper dictionary words, but stemming may create incomplete words. So lemmatization is mostly better for NLP

3.Preprocessing remove unwanted data and make the text more clean and consistent. This helps the NLP model understand the text better and improve the results
