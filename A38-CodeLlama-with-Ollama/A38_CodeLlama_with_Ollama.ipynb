```python
# Task 3 - Code Assistant Functions
#To make the code reusable, I created separate functions for each task supported by the coding assistant

def code_generation(prompt):
    response = llm.invoke(f"Write Python code for the following requirement:\n\n{prompt}")
    return response.content

def code_explanation(code):
    response = llm.invoke(f"Explain the following Python code in simple language:\n\n{code}")
    return response.content

def bug_fixing(code):
    response = llm.invoke(f"Find and fix the bugs in the following Python code. Also explain the corrections:\n\n{code}")
    return response.content

def code_optimization(code):
    response = llm.invoke(f"Optimize the following Python code for readability and performance:\n\n{code}")
    return response.content
```

```python
#testing
print("=== Code Generation ===\n")
result = code_generation("Write a Python function to reverse a string.")
print(result)
```

```python
print("=== Code Explanation ===\n")
sample_code = """
def add(a, b):
    return a + b

print(add(5, 10))
"""

result = code_explanation(sample_code)
print(result)
```

```python
#fixing the bug
print("=== Bug Fixing ===\n")

bugc = """
for i in range(5)
    print(i)
"""

result = bug_fixing(bugc)
print(result)
```

```python
#code optimization code
print("=== Code Optimization ===\n")
code = """
numbers = []

for i in range(100):
    numbers.append(i)
"""

result = code_optimization(code)
print(result)
```

All four functions produced useful results. CodeLlama was able to generate code, explain existing code, identify syntax errors, and suggest better implementations
