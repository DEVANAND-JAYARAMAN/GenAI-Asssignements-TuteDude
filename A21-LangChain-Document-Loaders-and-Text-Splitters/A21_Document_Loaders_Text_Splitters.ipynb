# Task 6: Why Text Splitting is Required

### 1. Why large documents cannot be directly passed to LLMs?

Large Language Models (LLMs) have a limited context window, meaning they can only process a certain amount of text at a time. Passing an entire large document may exceed this limit.

### 2. What problems does chunking solve?

Text chunking breaks large documents into smaller, manageable pieces. It improves retrieval accuracy, reduces memory usage, and helps preserve relevant information for downstream tasks like RAG and chatbots.

```python
#Task 7 – CharacterTextSplitter

from langchain_text_splitters import CharacterTextSplitter

print("="*60)
print("TASK 7 : CHARACTER TEXT SPLITTER")
print("="*60)

splitter = CharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50,
    separator="\n"
)

chunks = splitter.split_documents(text_documents)

print("Number of Chunks:", len(chunks))

print("\nFirst Chunk:\n")
print(chunks[0].page_content)

print("\nMetadata:")
print(chunks[0].metadata)
```

# Task 10: Semantic Chunking

### What is Semantic Chunking?

Semantic chunking groups text based on meaning rather than fixed character length. Related sentences are kept together to preserve context.

### How do embeddings help?

Embeddings represent the semantic meaning of text as vectors. By comparing embedding similarity, semantically related sentences can be grouped into the same chunk.

### Demo

Semantic chunking is commonly implemented using embedding models and semantic splitters. It produces more meaningful chunks than traditional character-based splitting, especially in Retrieval-Augmented Generation (RAG) systems.

```python
#Task 11 – Unified Preprocessing Pipeline

from langchain_community.document_loaders import (
    TextLoader,
    CSVLoader,
    PyPDFLoader,
    WebBaseLoader
)
from langchain_text_splitters import RecursiveCharacterTextSplitter
import os

def load_and_split_documents(path_or_url):
    """
    Loads documents from a local file, directory, or web URL,
    then splits them into smaller chunks.
    """
    documents = []

    # Handle Web URL
    if path_or_url.startswith("http"):
        loader = WebBaseLoader(path_or_url)
        documents = loader.load()

    # Handle Directory
    elif os.path.isdir(path_or_url):

        # Load TXT files
        txt_loader = DirectoryLoader(
            path_or_url,
            glob="**/*.txt",
            loader_cls=TextLoader
        )

        # Load CSV files
        csv_loader = DirectoryLoader(
            path_or_url,
            glob="**/*.csv",
            loader_cls=CSVLoader
        )

        # Load PDF files individually
        pdf_files = [
            file for file in os.listdir(path_or_url)
            if file.endswith(".pdf")]

        documents.extend(txt_loader.load())
        documents.extend(csv_loader.load())

        for pdf in pdf_files:
            pdf_loader = PyPDFLoader(os.path.join(path_or_url, pdf))
            documents.extend(pdf_loader.load())

    else:
        raise ValueError("Invalid Path or URL")

    splitter = RecursiveCharacterTextSplitter(chunk_size=300,chunk_overlap=50)
    chunks = splitter.split_documents(documents)
    return chunks
```

```python
#Apply on Local Directory
local_chunks = load_and_split_documents("/content/data")

print("Total Chunks:", len(local_chunks))

print("\nSample Chunk:\n")
print(local_chunks[0].page_content)

#Apply on Web URL

web_chunks = load_and_split_documents(
    "https://python.langchain.com/docs/introduction/"
)

print("Total Web Chunks:", len(web_chunks))

print("\nSample Chunk:\n")
print(web_chunks[0].page_content)
```
