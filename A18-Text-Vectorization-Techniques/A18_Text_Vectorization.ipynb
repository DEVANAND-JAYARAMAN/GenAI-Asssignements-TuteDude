```python
#Import Libraries

import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("/content/processed_sms_dataset.csv")

df.head()
```

```python
#Task 1: Manual One-Hot Encoding
sentences = df["final_clean_text"].head(5).tolist()   #selecting the 5 sentences

for i, sentence in enumerate(sentences, 1):
    print(f"Sentence {i}: {sentence}")


#build the vocabulary

vocabulary = sorted(set(word for sentence in sentences for word in sentence.split()))

print("Vocabulary:\n")
print(vocabulary)
print("\nVocabulary Size:", len(vocabulary))
```

```python
#Create One-Hot Vectors
one_hot_vectors = []

for sentence in sentences:
    vector = {}
    words = sentence.split()
    for word in vocabulary:
        vector[word] = 1 if word in words else 0
    one_hot_vectors.append(vector)

#Display One-Hot Matrix
one_hot_df = pd.DataFrame(one_hot_vectors)
one_hot_df

```

```python
#Task 2: One-Hot Encoding using Scikit-learn

binary_vectorizer = CountVectorizer(binary=True)
binary_matrix = binary_vectorizer.fit_transform(sentences)

#vocabulary
print("Vocabulary:\n")
print(binary_vectorizer.vocabulary_)

binary_df = pd.DataFrame(binary_matrix.toarray(),columns=binary_vectorizer.get_feature_names_out())
binary_df

```

### Observation

In manual one-hot encoding, each unique word in the vocabulary is assigned a binary value indicating whether it is present in a sentence. Using CountVectorizer(binary=True) produces the same result automatically and is more efficient for larger datasets.

```python
bow_vectorizer = CountVectorizer()
# Fit and transform the complete dataset after filling NaN values with empty strings
bow_matrix = bow_vectorizer.fit_transform(df["final_clean_text"].fillna(''))
vocabulary = bow_vectorizer.get_feature_names_out()
print("Vocabulary Size:", len(vocabulary))

#display te 20words
print("First 20 Words in Vocabulary:\n")
print(vocabulary[:20])

bow_df = pd.DataFrame(bow_matrix.toarray(),columns=bow_vectorizer.get_feature_names_out())
bow_df.head()

print("Sample Feature Vectors:\n")
bow_df.head(5)
```

```python
#Task 4: Understanding Word Frequency

word_frequency = bow_df.sum(axis=0)
frequency_df = pd.DataFrame({"Word": word_frequency.index,"Frequency": word_frequency.values})
frequency_df.head()

top_words = frequency_df.sort_values(by="Frequency",ascending=False).head(10)
top_words

least_words = frequency_df.sort_values(by="Frequency").head(10)
least_words


```

## Observation

The Bag of Words model represents each document by counting the number of times each word appears. Frequently occurring words receive higher counts, while words that appear rarely have lower counts. This representation captures word frequency but does not preserve the order or context of words.

```python
#Task 5: Unigrams, Bigrams & Trigrams

# Unigram Vectorizer
unigram_vectorizer = CountVectorizer(ngram_range=(1, 1))

unigram_matrix = unigram_vectorizer.fit_transform(df["final_clean_text"].fillna(''))

print("Unigram Vocabulary Size:", len(unigram_vectorizer.get_feature_names_out()))
print()

unigram_df = pd.DataFrame(
    unigram_matrix.toarray(),
    columns=unigram_vectorizer.get_feature_names_out()
)
unigram_df.head()

# Bigram Vectorizer
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram_vectorizer.fit_transform(df["final_clean_text"].fillna(''))

print("Bigram Vocabulary Size:", len(bigram_vectorizer.get_feature_names_out()))
print()

bigram_df = pd.DataFrame(
    bigram_matrix.toarray(),
    columns=bigram_vectorizer.get_feature_names_out()
)
bigram_df.head()

# Trigram Vectorizer
trigram_vectorizer = CountVectorizer(ngram_range=(3, 3))
trigram_matrix = trigram_vectorizer.fit_transform(df["final_clean_text"].fillna(''))

print("Trigram Vocabulary Size:", len(trigram_vectorizer.get_feature_names_out()))
print()

trigram_df = pd.DataFrame(
    trigram_matrix.toarray(),
    columns=trigram_vectorizer.get_feature_names_out()
)
trigram_df.head()

comparison = pd.DataFrame({
    "Vectorizer": ["Unigram", "Bigram", "Trigram"],
    "Vocabulary Size": [
        len(unigram_vectorizer.get_feature_names_out()),
        len(bigram_vectorizer.get_feature_names_out()),
        len(trigram_vectorizer.get_feature_names_out())]})
comparison
```

```python
#Task 6: Combined N-Grams

combined_vectorizer = CountVectorizer(ngram_range=(1, 2))
combined_matrix = combined_vectorizer.fit_transform(df["final_clean_text"].fillna(''))

print("Combined Vocabulary Size:",
      len(combined_vectorizer.get_feature_names_out()))

combined_df = pd.DataFrame(
    combined_matrix.toarray(),
    columns=combined_vectorizer.get_feature_names_out()
)
combined_df.head()
```

## Observation

- Unigrams represent individual words.
- Bigrams capture relationships between two consecutive words.
- Trigrams preserve even more context by considering three consecutive words.
- Combining unigrams and bigrams provides richer contextual information while still retaining individual word frequencies.

```python
#Task 7: TF-IDF Implementation

# Create TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()
# Fit and transform the dataset
tfidf_matrix = tfidf_vectorizer.fit_transform(df["final_clean_text"].fillna(''))

print("Vocabulary Size:", len(tfidf_vectorizer.get_feature_names_out()))
print("\nFirst 20 Vocabulary Words:\n")
print(tfidf_vectorizer.get_feature_names_out()[:20])
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),columns=tfidf_vectorizer.get_feature_names_out())

tfidf_df.head()
```

```python
#Task 8: BoW vs TF-IDF Comparison

comparison_words = ["free", "call", "please", "win", "love"]
comparison_df = pd.DataFrame({"BoW Frequency": bow_df[comparison_words].sum(),"TF-IDF Weight": tfidf_df[comparison_words].mean()})
comparison_df

average_scores = tfidf_df.mean(axis=0)
high_tfidf = average_scores.sort_values(ascending=False).head(10)
high_tfidf

low_tfidf = average_scores.sort_values().head(10)
low_tfidf

```

## Observation

Bag of Words stores only the frequency of words in each document. TF-IDF assigns higher importance to words that appear frequently in a document but rarely across the entire dataset. As a result, common words receive lower weights while informative words receive higher weights.

```python
#Task 9: Vectorizer Parameter Exploration
vectorizer_max = CountVectorizer(max_features=100)
vectorizer_max.fit(df["final_clean_text"].fillna(''))

print("Vocabulary Size with max_features=100:")
print(len(vectorizer_max.get_feature_names_out()))

vectorizer_min = CountVectorizer(min_df=2)
vectorizer_min.fit(df["final_clean_text"].fillna(''))

print("Vocabulary Size with min_df=2:")
print(len(vectorizer_min.get_feature_names_out()))

vectorizer_maxdf = CountVectorizer(max_df=0.8)
vectorizer_maxdf.fit(df["final_clean_text"].fillna(''))

print("Vocabulary Size with max_df=0.8:")
print(len(vectorizer_maxdf.get_feature_names_out()))
```

```python
parameter_df = pd.DataFrame({
    "Parameter": ["Default","max_features=100","min_df=2","max_df=0.8"],
    "Vocabulary Size": [
        len(bow_vectorizer.get_feature_names_out()),
        len(vectorizer_max.get_feature_names_out()),
        len(vectorizer_min.get_feature_names_out()),
        len(vectorizer_maxdf.get_feature_names_out())]})
parameter_df
```

1. One-Hot Encoding represents only whether a word is present or absent in a sentence. Bag of Words represents the number of times each word appears in a document

2. N-Grams generate combinations of consecutive words, creating many additional features. As the value of N increases, the vocabulary grows significantly, resulting in higher-dimensional feature vectors

3. TF-IDF is preferred when common words should have less influence and important words should receive higher importance. It often performs better in text classification and information retrieval tasks

4. Ignores word meaning.
- Ignores word order (except N-Grams).
- Produces sparse vectors.
- Cannot understand context or semantic relationships
